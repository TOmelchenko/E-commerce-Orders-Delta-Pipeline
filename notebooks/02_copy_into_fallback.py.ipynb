```python
%sql
DROP TABLE IF EXISTS retail_dev.orders.bronze_orders_stg;
CREATE TABLE IF NOT EXISTS retail_dev.orders.bronze_orders_stg(
  order_id STRING,
  customer_id STRING,
  order_status STRING,
  order_amount STRING,
  order_date STRING,
  shipping_method STRING,
  payment_status STRING,
  delivery_country STRING,
  last_updated_ts STRING
)
USING DELTA; -- that makes table delta table, not odinary table

CREATE TABLE IF NOT EXISTS retail_dev.orders.bronze_orders_into(
  order_id STRING,
  customer_id STRING,
  order_status STRING,
  order_amount DOUBLE,
  order_date DATE,
  shipping_method STRING,
  payment_status STRING,
  delivery_country STRING,
  last_updated_ts TIMESTAMP
)
USING DELTA; 
```

```python
%sql
COPY INTO retail_dev.orders.bronze_orders_stg
FROM '/Volumes/retail_dev/orders/landing/orders'
FILEFORMAT = CSV
FORMAT_OPTIONS (
  'header' = 'true'
);
```

```python
%sql
INSERT INTO retail_dev.orders.bronze_orders_into
SELECT
  order_id,
  customer_id,
  order_status,
  CAST(order_amount AS DOUBLE) as order_amount,
  CAST(order_date AS DATE) as order_date, 
  shipping_method,
  payment_status,
  delivery_country,
  CAST(last_updated_ts AS TIMESTAMP) as last_updated_ts
FROM retail_dev.orders.bronze_orders_stg
```
